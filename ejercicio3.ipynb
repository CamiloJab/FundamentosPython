Un vendedor recibe un sueldo base más un 10% extra por
comisión de sus ventas, el vendedor desea saber cuánto
dinero obtendrá por concepto de comisiones por las tres
ventas que realiza en el mes y el total que recibirá en
el mes tomando en cuenta su sueldo base y comisiones.

```python
venta1 = float(input("Por favor indique sus venta # 1: "))
venta2 = float(input("Por favor indique sus venta # 2: "))
venta3 = float(input("Por favor indique sus venta # 3: "))

sueldo =((venta1 + venta2 + venta3)*0.10)

print(f"El calculo de su sueldo sera por un valor de = {sueldo} ")
```

Una tienda ofrece un descuento del 15% sobre el total
de la compra y un cliente desea saber cuánto deberá
pagar finalmente por su compra.

```python
TotalCompra = int(input("Indique el valor total de su compra: "))

descuento = (TotalCompra*0.15)
Pagar = (TotalCompra - descuento)
print(f"El valor total a pagar es de = {Pagar}")
```

Un alumno desea saber cuál será su calificación final en
la materia de Algoritmos. Dicha calificación se compone
de tres exámenes parciales.

```python
nota1 = float(input("Digita la nota del primer trimestre "))
nota2 = float(input("Digita la nota del segundo trimestre "))
nota3 = float(input("Digita la nota del tercer trimestre "))

notaFinal = round((nota1 + nota2 + nota3)/3, 2)

print(f"Su notal final de la materia es : {notaFinal}")
```

Un maestro desea saber qué porcentaje de hombres y que
porcentaje de mujeres hay en un grupo de estudiantes.

```python
totalAlumnos = int(input("Favor indicar el total de alumnos en la clase"))
mujeres = int(input("Favor indique cuantas mujeres tiene en clase "))

porMujeres = ((mujeres/totalAlumnos)*100)
porHombres = ((porMujeres-100)*-1)

print(f"El porcentaje de mujeres en clase es de {porMujeres} y de hombres es de {porHombres}")
```
